```python

last_shared_task = None

def f(number):
    return "{0:.5f}".format(number)

def print_header(evaluation):
    ret = ''
    import statistics
    global last_shared_task
    if last_shared_task is None or last_shared_task != evaluation['SharedTask']:
        last_shared_task = evaluation['SharedTask']
        ret += '\n\n# ' + last_shared_task + '\n\n' +\
               '|Evaluation|Dedup|QRELs|Preprocessing' + \
               '|AVG(score)|AVG(Man-Score)|Diff-Scores' +\
               '|Max(ScoreChanges)|Median(ScoreChanges)|Min(ScoreChanges)' + \
               '|Max(Ideal)|Median(Ideal)|Min(Ideal)' +\
               '|Max(Cheat)|Median(Cheat)|Min(Cheat)|Kendall-Tau|Kendall-Tau@5|Kenbdall-Tau@10|\n' +\
               '|:---------|:------------|:----|:------------' + \
               '|---------:|-------------:|----------:' +\
               '|---------:|----------------:|---------------:' + \
               '|---------:|------------:|--------:' +\
               '|---------:|------------:|--------:|----------:|--------------:|-------------:|\n'
    scoreChanges = evaluation['ScoreChangesPerRank']
    ideal = evaluation['IdealParticipantVector']
    cheat = evaluation['CheatingParticipantVector']
    avgEval = evaluation['AverageRankEval']
    
    line = '|' + evaluation['EvaluationMeasure'] + '|' + evaluation['Deduplication'].replace('Duplicates','') + '|' + evaluation['QrelConsistency'] + '|' + evaluation['SystemRankingPreprocessing'] + \
               '|' + f(avgEval['avgOriginal']) + '|' + f(avgEval['avgManipulated']) + '|' + f(avgEval['difference']) +\
               '|' + f(max(scoreChanges)) + '|' + f(statistics.median(scoreChanges)) + '|' + f(min(scoreChanges)) + \
               '|' + str(max(ideal)) + '|' + str(statistics.median(ideal)) + '|' + str(min(ideal)) +\
               '|' + str(max(cheat)) + '|' + str(statistics.median(cheat)) + '|' + str(min(cheat)) +\
               '|' + f(evaluation['KendallTau']) + '|' + f(evaluation['KendallTau@5']) +'|' + f(evaluation['KendallTau@10']) + '|\n'
    
    #if evaluation['QrelConsistency'] == 'base':
    ret += line
    
    return ret

def kpis(vector):
    import statistics
    return '\n  -  Max: ' + str(max(vector)) +\
            '\n  -  Mean: ' + str(statistics.median(vector))

def sort_evals(evaluations):
    return sorted(sorted(
        sorted(evaluations, key=lambda i: i['EvaluationMeasure']),
        key= lambda i: i['SystemRankingPreprocessing']
    ), key= lambda i: i['SharedTask'])


def bar_chart(vector, plot_name):
    import matplotlib.pyplot as plt
    from IPython.display import Image
    
    fig = plt.figure()
    plt.scatter(x=range(len(vector)), y=[i for i in vector])
    
    plt.savefig(plot_name)
    plt.close()
    return Image(plot_name + '.png')

scoreChanges = []
cheatingVector = []
idealVector = []


def print_report(evaluation):
    name = evaluation['QrelConsistency'] + evaluation['Deduplication'] + evaluation['EvaluationMeasure']
    global scoreChanges
    global cheatingVector
    global idealVector
    
    
    
    scoreChanges += [bar_chart(evaluation['ScoreChangesPerRank'], name+'-sc')]
    cheatingVector += [bar_chart(evaluation['CheatingParticipantVector'], name + '-cv')]
    idealVector += [bar_chart(evaluation['IdealParticipantVector'], name + '-iv')]
    
    return ''
    #return '-  Kendall-Tau: ' + str(evaluation['KendallTau']) +\
    #        '\n-  Score-Changes-Per-Rank' + kpis(evaluation['ScoreChangesPerRank']) \
    #            + '\n {{imgs[\'' + name + '-score-changes\']}} '+\
    #        '\n-  Cheating-Vector' + kpis(evaluation['CheatingParticipantVector']) +\
    #        '\n-  Ideal-Vector\n' + kpis(evaluation['IdealParticipantVector'])



def report():
    from IPython.display import display, Markdown, Latex
    global last_shared_task
    last_shared_task = None
    
    global scoreChanges
    scoreChanges = []
    global cheatingVector
    cheatingVector = []
    global idealVector
    idealVector = []
    
    ret = ''
    for i in sort_evals(evaluation_data()):
        ret += print_header(i)
        print_report(i)
    display(Markdown(ret))
```
